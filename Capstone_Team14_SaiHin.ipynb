```python
# os = operating system
import os
import pandas as pd
import matplotlib.pyplot as plt

print("Path at terminal when executing this file")
print(os.getcwd())
```

```python
path_whr2017 = "data/coronavirusdataset/whr2017.csv"
path_whr2018 = "data/coronavirusdataset/whr2018.csv"

whr2017 = pd.read_csv(path_whr2017)
whr2018 = pd.read_csv(path_whr2018)
```
